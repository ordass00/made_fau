### Datasource 2: HadCRUT 5 (Temperature Dataset)

```python
hadcrut5_data_url = "https://crudata.uea.ac.uk/cru/data/temperature/HadCRUT.5.0.2.0.analysis.anomalies.ensemble_mean.nc#mode=bytes"

hadcrut5_data = Dataset(hadcrut5_data_url)
print(hadcrut5_data)
```

```python
hadcrut5_data.variables.keys()
```

```python
print(hadcrut5_data.variables["time"])
```

```python
print(hadcrut5_data.variables["tas_mean"])
```

```python
print(hadcrut5_data.variables["latitude"])
```

```python
print(hadcrut5_data.variables["longitude"])
```

```python
time = hadcrut5_data.variables["time"][:]
time
```

```python
time_units = hadcrut5_data.variables["time"].units
time_units
```

```python
dates = cftime.num2date(time, units=time_units, calendar="standard")
dates
```

```python
time = pd.to_datetime([date.strftime("%Y-%m-%d %H:%M:%S") for date in dates])
time
```

```python
latitude = hadcrut5_data.variables["latitude"][:]
latitude
```

```python
longitude = hadcrut5_data.variables["longitude"][:]
longitude
```

```python
temperature_anomaly = hadcrut5_data.variables["tas_mean"][:]
temperature_anomaly
```

```python
time_repeated = np.repeat(time, len(latitude) * len(longitude))
latitudes_tiled = np.tile(np.repeat(latitude, len(longitude)), len(time))
longitudes_tiled = np.tile(longitude, len(time) * len(latitude))
tas_mean_flattened = temperature_anomaly.flatten()
```

```python
temperature_df = pd.DataFrame({
    "time": time_repeated,
    "latitude": latitudes_tiled,
    "longitude": longitudes_tiled,
    "temperature_anomaly": tas_mean_flattened
})
```

```python
temperature_df.head(20)
```

```python
temperature_df.describe()
```

#### Example output for Germany: Latitude from round about 47.0 to 55.0 and Longitude round about from 6 to 15.0

```python
latitude_min, latitude_max = 47.0, 55.0
lonitude_min, lonitude_max = 5.0, 15.0

germany_temperature_df = temperature_df[
    (temperature_df["latitude"] >= latitude_min) & (temperature_df["latitude"] <= latitude_max) &
    (temperature_df["longitude"] >= lonitude_min) & (temperature_df["longitude"] <= lonitude_max)
]
```
